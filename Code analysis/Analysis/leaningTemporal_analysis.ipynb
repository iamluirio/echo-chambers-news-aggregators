# Leaning Temporal Graph Analysis

Analysis carried out over 5 days: the progress of the topic over the various days is evaluated, calculating the number of news provided by google news, for each day, for each topic and for each user.

1. 4 graphs for the USA, 2 evaluate the pro-republican and 2 pro-democratic trend.

2. 4 graphs for India politics, 2 evaluate the pro-gov and 2 the pro-opp trend.

3. 4 graphs for Macro/Micro topics, 2 evaluate the macro and 2 the micro trend.

## USA

### (Pier): Pro-republican news rate.

```python
import matplotlib.pyplot as plt
import numpy as np

# Set up data for the graph
days = ['02/04', '03/04', '04/04', '05/04', '06/04']
users = ['U1', 'U2', 'U3', 'U4']
max_rate = 6
news_rates = np.random.randint(1, max_rate+1, size=(len(days), len(users)))
news_rates[:, 0] = [3, 0, 2, 1, 2]  # Set the values for U1
news_rates[:, 1] = [3, 2, 4, 0, 3]  # Set the values for U2
news_rates[:, 2] = [0, 1, 0, 1, 1]  # Set the values for U3
news_rates[:, 3] = [0, 0, 0, 3, 0]  # Set the values for U4
shapes = ['o', 's', 'D', '^', '*', 'x', 'p', 'h']  # List of geometric shapes to use for each point
colors = ['red' if i == 0 else 'brown' if i == 1 else 'blue' if i == 2 else 'green' if i == 3 else None for i in range(len(users))]  # Set the color of the lines

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))  # Set the figure size to 12x8 inches
handles = []
line_styles = ['-', ':', '--', '-.'] 
for i, user in enumerate(users):
    line_style = line_styles[i]  # Select the appropriate line style for the user
    line, = ax.plot(days, news_rates[:, i], color=colors[i], linestyle=line_style)
    marker = ax.scatter(days, news_rates[:, i], marker=shapes[i % len(shapes)], color=colors[i] if i != 0 else 'red')
    handles.append((line, marker))
    ax.plot([], [], color=line.get_color(), linestyle=line_style, label=user)

# Set axis labels, limits, and ticks
ax.set_xlabel('Date')
ax.set_ylabel('News Articles')
ax.set_ylim([0, max_rate])
ax.set_yticks(np.arange(max_rate+1, dtype=int))

# Set legend 
ax.legend(handles=handles, labels=users, loc='upper right', fontsize=12, prop={'size': 10}, handlelength=2.5, handleheight=1.5)

# Saving as eps plot image
plt.savefig('pier_proRep.eps', format='eps')

# Display the plot
plt.show()
```

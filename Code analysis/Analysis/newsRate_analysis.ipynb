# News Rate Analysis

## Analysys carried out on different days (separately) on top 20 news for each day, for each user, and plotted a multi-bar graph for each topic.

### Tuesday 21 March (Pier): Users 1 to 4

```python
import matplotlib.pyplot as plt
import numpy as np

# Set data for the plot
REP = [3, 1, 2, 2]
DEM = [6, 5, 8, 4]
UNB = [0, 0, 3, 7]
MIC = [2, 5, 1, 3]
OTH = [9, 9, 6, 4]
x = np.arange(len(REP))  # x-axis values

# Set bar width
bar_width = 0.15

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))  # set the figure size to 12x8
colormap = plt.get_cmap('Blues')  # Choose the colormap
colors = [colormap(i) for i in np.linspace(0.2, 1, 5)]  # Generate 5 colors from the colormap starting at 0.2

ax.bar(x - 2.0 * bar_width, REP, bar_width, label='rep', color=colors[0])
ax.bar(x - 1.0 * bar_width, DEM, bar_width, label='dem', color=colors[1])
ax.bar(x, UNB, bar_width, label='unbiased', color=colors[2])
ax.bar(x + 1.0 * bar_width, MIC, bar_width, label='micro', color=colors[3])
ax.bar(x + 2.0 * bar_width, OTH, bar_width, label='others', color=colors[4])

# Set the x-axis labels and ticks
ax.set_xticks(x)
ax.set_xticklabels(['U1', 'U2', 'U3', 'U4'])
ax.set_xlabel('Users', fontsize=14)
ax.set_ylabel('Num of News Articles', fontsize=14)

# Add a legend and title
ax.legend()

# Saving as eps plot image
plt.savefig('plots/tuesday21_pier.eps', format='eps')

# Show the plot
plt.show()
```

### Wednesday 29 March (Ananya B.): Users 5 to 8

```python
import matplotlib.pyplot as plt
import numpy as np

# Set data for the plot
MAC = [16, 19, 17, 9]
MIC = [4, 1, 3, 11]
OTH = [0, 0, 0, 0]
x = np.arange(len(MAC))  # x-axis values

# Set bar width
bar_width = 0.15

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))  # set the figure size to 12x8
colormap = plt.get_cmap('Blues')  # Choose the colormap
colors = [colormap(i) for i in np.linspace(0.2, 1, 3)]  # Generate 3 colors from the colormap starting at 0.2

ax.bar(x - bar_width, MAC, bar_width, label='macro', color=colors[0])
ax.bar(x, MIC, bar_width, label='micro', color=colors[1])
ax.bar(x + bar_width, OTH, bar_width, label='others', color=colors[2])

# Set the x-axis labels and ticks
ax.set_xticks(x)
ax.set_xticklabels(['I7', 'I8', 'I9', 'I10'])
ax.set_xlabel('Users', fontsize=14)
ax.set_ylabel('Num of News Articles', fontsize=14)

# Set y-axis ticks
yticks = range(0, max(MAC + MIC + OTH) + 1, 2)  # set y-axis ticks to be integers
ax.set_yticks(yticks)

# Set y-axis tick labels
yticklabels = [str(y) if y % 2 == 0 else '' for y in yticks]  # set even numbers as tick labels
ax.set_yticklabels(yticklabels)

# Add a legend and title
ax.legend()

# Saving as eps plot image
plt.savefig('wednesday29_ananyaB.eps', format='eps')

# Show the plot
plt.show()
```

# Yahoo News Aggregator

```python
import requests
import xmltodict

# Set the API endpoint URL and parameters
url = 'https://news.yahoo.com/rss/'
params = {'region': 'US', 'count': 10}

# Make the API request and get the response data
response = requests.get(url, params=params)
data = response.text

# Parse the XML data into a Python dictionary
xml_data = xmltodict.parse(data)

# Extract the news article titles and URLs
news_articles = []
for item in xml_data['rss']['channel']['item']:
    title = item['title']
    url = item['link']
    news_articles.append((title, url))

# Display the news articles
for title, url in news_articles:
    print(title)
    print(url)
    print()
```

# Positional Homophily in News Recommendation

As Google News ranks the news articles based on relevance to the user, we explore the positional homophily in news recommendation for an user, i.e., we study the first position of any news article that belong to the userâ€™s macro news topic.

An user has high positional homophily in news recommendation if the news articles which match her macro news topic is ranked early in her news feed. For our experiment, we repeat this for all the users of both USAand India for 5 days, respectively.

### India

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set data for the plot
firstDay = [1, 1, 2, 17, 1, 4]
secondDay = [2, 1, 2, 3, 2, 2]
thirdDay = [1, 1, 1, 1, 1, 1]
x = np.arange(len(firstDay))  # x-axis values

# Set bar width
bar_width = 0.15

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))  # set the figure size to 12x8
colormap = plt.get_cmap('Blues')  # Choose the colormap
colors = [colormap(i) for i in np.linspace(0.2, 1, 3)][::-1]  # Generate 3 colors from the colormap starting at 0.2 and reverse the order

ax.bar(x - bar_width, firstDay, bar_width, label='12/04', color=colors[0])
ax.bar(x, secondDay, bar_width, label='13/04', color=colors[1])
ax.bar(x + bar_width, thirdDay, bar_width, label='14/04', color=colors[2])

# Set the x-axis labels and ticks
ax.set_xticks(x)
ax.set_xticklabels(['I1', 'I2', 'I3', 'I4', 'I5', 'I6'])
ax.set_xlabel('Users', fontsize=14)
ax.set_ylabel('Position', fontsize=14)

# Set the legend position
ax.legend(loc='upper right')

# Set only natural numbers in the Y-axis
ax.set_ylim([0, 18])
ax.yaxis.set_ticks(np.arange(0, 19, 1))

# Saving as eps plot image
#plt.savefig('fig_8_a.eps', format='eps')

# Show the plot
plt.show()
```
